#### Q.Plot artist wise mean of energy and liveness on a bar chart for the artist who have 3 or more songs 

## 1) . Bar chart

```python
artist_3 = df['artist'].value_counts()
artist_3 [artist_3 >= 3]
r1 = artist_3 [artist_3 >= 3]
r2 = r1.index
```

```python
res1= df[df['artist'].isin (r2)].groupby('artist')[['energy','liveness']].mean()
res1
```

```python
res1.plot(kind = 'bar')
plt.show()
```

```python
res1.plot.bar()
plt.show()
```

```python
res1.plot.bar(stacked = True)
plt.grid()
plt.show()
```

```python
res1.plot.bar(subplots = True)
plt.show()
```

```python
res1.plot.bar(subplots = True, color = {'energy' : 'maroon',
                                      'liveness' : 'pink'})
plt.show()
```

### 3) Pie chart

####  Q. plot year wise mean of popularity on a pie chart for the years 2017 and above:

```python
df_year2017 = df[df['year'] >= 2017]
df_year2017
```

```python
r5 = df_year2017.groupby('year')['popularity'].mean()
r5
```

```python
r5.plot.pie(autopct = '%.2f%%')
plt.show()
```

#### Q.Plot the year 2017 and more wise popularity and energy using pie chart:


```python
r6 = df_year2017.groupby('year')[['popularity', 'energy']].mean()
r6
```

```python
r6.plot.pie(subplots = True , autopct = '%.2f%%')
r6
```
